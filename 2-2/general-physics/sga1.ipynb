```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.01
T = 5
N = int(T/dt)
b = 0.02
g = np.array([0, -9.8])
q_0 = np.array([0, 0])
v_0 = np.array([100, 100])
```

```python
q_t = np.zeros((2, N))
q = q_0
v = v_0
for i in range(N):
    a = g - b*np.linalg.norm(v)*v
    v = v + a*dt
    q = q + v*dt
    q_t[:, i] = q
plt.plot(q_t[0, :], q_t[1, :])
plt.show()
```
