```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
dt = 0.01
T = 5
N = int(T/dt)
b = 0.02
g = np.array([0, -9.8])
q_0 = np.array([0, 0])
v_0 = 50
theta = np.pi/3
```

```python
q_t = np.zeros((2, N))
q_t[:, 0] = q_0
v = np.array([v_0*np.cos(theta), v_0*np.sin(theta)])
for i in range(1,N):
    q_t[:, i] = q_t[:, i-1] + v*dt
    a = g - b*np.linalg.norm(v)*v
    v = v + a*dt
plt.plot(q_t[0, :], q_t[1, :])
plt.show()
```
